```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Формирование дат начала и конца периода
end_date = datetime.now()
start_date = end_date - timedelta(days=50)

# Преобразование дат в строковый формат
start_date_str = start_date.strftime('%Y-%m-%d')
end_date_str = end_date.strftime('%Y-%m-%d')

# Инициализация переменных для цикла
start_index = 0
batch_size = 500  # предполагаемое ограничение количества записей на один запрос
all_candles_data = []  # список для хранения всех полученных данных

while True:
    # URL для запроса данных с учетом start_index
    url = f'http://iss.moex.com/iss/engines/stock/markets/shares/boards/TQBR/securities/SBER/candles.json?from={start_date_str}&till={end_date_str}&interval=1&start={start_index}'

    # Выполнение запроса
    response = requests.get(url)
    data = response.json()

    # Извлечение данных свечек
    candles_data = data['candles']['data']

    # Проверка наличия данных в ответе
    if not candles_data:
        break  # Выход из цикла, если данных больше нет

    # Добавление полученных данных в общий список
    all_candles_data.extend(candles_data)

    # Увеличение start_index для следующего запроса
    start_index += batch_size

# Создание DataFrame из всех полученных данных
df = pd.DataFrame(all_candles_data, columns=data['candles']['columns'])

# Преобразование типов данных для даты и времени
df['begin'] = pd.to_datetime(df['begin'])

df.info()  # Вывод первых строк DataFrame для проверки
```

```python
# Добавление новых колонок
df['month'] = df['begin'].dt.month
df['day_of_week'] = df['begin'].dt.day_name()  # или df['begin'].dt.weekday для числового представления
df['hour'] = df['begin'].dt.hour

# Удаляем ненужные колонки, оставляем только 'close' и другие неценовые колонки
df = df[['close', 'volume', 'begin', 'month', 'day_of_week', 'hour']]


# Добавляем 100 предыдущих значений 'close' как новые колонки
for i in range(1, 101):
    df[f'close_{i}'] = df['close'].shift(i)


df['target'] = df['close'].shift(-60)

# Удаляем строки, где появляются NaN из-за сдвига
df = df.dropna().reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Определение категориальных и числовых признаков
categorical_features = ['day_of_week', 'month', 'hour']
numeric_features = [f'close_{i}' for i in range(1, 101)]  

# Создание трансформера для категориальных признаков
categorical_transformer = OneHotEncoder()

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Создание ColumnTransformer для применения OHE только к категориальным колонкам
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
       ('cat', categorical_transformer, categorical_features)
    ])

# Создание pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000000))
])

# Процентное изменение между close_1 и target
df['percent_change'] = ((df['target'] - df['close_1']) / df['close_1']) * 100

    # Порог изменения курса
x = 0.2  # 1%

# Создание бинарной целевой переменной
df['target_class'] = (df['percent_change'] >= x).astype(int)

    # Определение признаков X и целевой переменной y
X = df[numeric_features + categorical_features]
y = df['target_class']

    # Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)



# Обучаем модель на нормализованных данных
pipeline.fit(X_train, y_train)

    # Предсказание на тестовой выборке
y_pred = pipeline.predict(X_test)

    # Расчет точности
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc

# предположим, что model - ваша обученная модель
# X_test, y_test - тестовые признаки и метки

# Получаем предсказанные вероятности
y_scores = pipeline.predict_proba(X_test)[:, 1]

# Рассчитываем ROC AUC
fpr, tpr, thresholds_roc = roc_curve(y_test, y_scores)
roc_auc = roc_auc_score(y_test, y_scores)

# Рассчитываем Precision-Recall AUC
precision, recall, thresholds_pr = precision_recall_curve(y_test, y_scores)
pr_auc = auc(recall, precision)

# Построение ROC Curve
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # Два графика рядом
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

# Построение Precision-Recall Curve
plt.subplot(1, 2, 2)
plt.plot(recall, precision, color='green', lw=2, label='PR curve (area = %0.2f)' % pr_auc)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")

plt.show()
```

```python
import matplotlib.pyplot as plt

def simulate_trading(TestingDF):
    X = TestingDF['close']
    y = TestingDF['pred']
    
    capital = 10000  # начальный капитал
    capital_curve = [capital]  # кривая капитала
    in_position = False  # индикатор наличия открытой позиции
    buy_price = 0  # цена покупки
    
    for i in range(len(X) - 1):
        if y[i] == 1 and not in_position:
            # Открываем позицию
            buy_price = X[i + 1]
            in_position = True
        elif in_position:
            change = (X[i + 1] - buy_price) / buy_price
            if change <= -0.0015:
                # Стоп-лосс
                capital *= (1 + change)
                in_position = False
            elif change >= 0.009:
                # Фиксация прибыли
                capital *= (1 + change)
                in_position = False
        
        capital_curve.append(capital)
    
    if in_position:
        # Закрываем последнюю позицию по последней известной цене
        change = (X[-1] - buy_price) / buy_price
        capital *= (1 + change)
        capital_curve.append(capital)

    plt.plot(capital_curve)
    plt.title("Кривая капитала")
    plt.xlabel("Шаги")
    plt.ylabel("Капитал")
    plt.show()
    return capital_curve
```
